```python
#BEER
# gaussianBlur : filtro de média com convolução
os.getcwd()
Pasta_Beer="/content/TRABALHO-PDI-CERTO/FOTOS/BEER"
Beerlen=len(Pasta_Beer)
dir1="/content/TRABALHO-PDI-CERTO/FOTOS/"
dir2="BEER"
dir3=dir1+dir2
seq=iaa.Sequential([
iaa.GaussianBlur(sigma=(0, 3.0)),
])
for img in glob.glob(Pasta_Beer + "/*.png"):
  filename = img.split('BEER')[-1][:-4] # pega o nome do arquivo
  for i in range(1):
    image = cv2.imread(img)
    images_aug = seq(images=[image])
    cv2.imwrite(dir3+'{}_{}.png'.format(filename,i),images_aug[0])

#Filtro exponencial FILTROBILATERAL

seq1=iaa.Sequential(
    iaa.BilateralBlur( d=(3, 10), sigma_color=(60, 250), sigma_space=(60, 250))
)
for img1 in glob.glob(Pasta_Beer + "/*.png"):
  filename1 = img1.split('BEER')[-1][:-4] # pega o nome do arquivo
  for i in (i* 0.5 for i in range(6)):
    image1 = cv2.imread(img1)
    images_aug1 = seq1(images=[image1])
    cv2.imwrite(dir3+'{}_{}.png'.format(filename,i),images_aug1[0])
# FILTRP EXTRA DE ROTAÇÃO EM 45°

    seq2=iaa.Sequential(
    iaa.Rotate((-45, 45))
)
for img2 in glob.glob(Pasta_Beer + "/*.png"):
  filename2 = img2.split('BEER')[-1][:-4] # pega o nome do arquivo
  for i in (i* 0.1 for i in range(3)):
    image2 = cv2.imread(img2)
    images_aug2 = seq2(images=[image2])
    cv2.imwrite(dir3+'{}_{}.png'.format(filename,i),images_aug2[0])

# LOGARITMO

image = cv2.imread('/content/TRABALHO-PDI-CERTO/FOTOS/BEER/0-01-V2-W.png')
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image2 = cv2.imread('/content/TRABALHO-PDI-CERTO/FOTOS/BEER/0-01-V1-B.png')
image2RGB = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

c1 = 255 / (np.log(1 + np.max(imageRGB))+0.00001)
log_image1 = c1 * np.log(imageRGB + 1  +0.00001)
log_image1 = np.array(log_image1, dtype = np.uint8)
filename = img.split('BEER')[-1][:-4] # pega o nome do arquivo
cv2.imwrite(dir3+'{}_{}.png'.format(filename,99),log_image1)

c2 = 255 /(np.log(1 + np.max(image2RGB))+0.00001)
log_image2 = c2 * (np.log(image2RGB + 1  + 0.00001))
log_image2 = np.array(log_image2, dtype = np.uint8)
filename1 = img1.split('BEER')[-1][:-4] # pega o nome do arquivo
cv2.imwrite(dir3+'{}_{}.png'.format(filename1,999),log_image2)

###################PLOTANDO AS IMAGENS#############################
#plotando todas as imagens por classe

#colocando as imagens da pasta em uma lista
images = []
for f in glob.glob('/content/TRABALHO-PDI-CERTO/FOTOS/BEER/*'):
  images.append(mpimg.imread(f))
plt.figure(figsize=(10,10))
coluna=5
for i, image in enumerate(images):
  plt.subplot(int(len(images)/coluna+1), coluna, i+1)
  plt.imshow(image)

```
